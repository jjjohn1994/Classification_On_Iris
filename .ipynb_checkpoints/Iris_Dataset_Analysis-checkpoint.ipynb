```python
import numpy as np
from sklearn import datasets
iris = datasets.load_iris()
X = iris.data[:, [2, 3]]
y = iris.target
print('Class labels:', np.unique(y))
```

Class labels:
0 - Iris-setosa
1 - Iris-vericolor
2 - Iris-virginica

```python
# To evaluate how well a trained model performs on unseen data,
# We will split the dataset into seperate training & test datasets
# Stratification - returns training & testing subsets that've the same
#                  proportions of class labels as the input dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,
                                                   random_state=1, stratify=y)
```

```python
# Using numpy's bincount() to count the num of occurences of each value
# in an array
print('Labels counts in y:', np.bincount(y))
```

```python
print('Labels count in y_train:', np.bincount(y_train))
```

```python
print('Labels count in y_test:', np.bincount(y_test))
```

```python
# Feature scaling for optimal performace
# Here we will standardize the features using the StandardScaler
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)
X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)
```

Using fit method, StandardScaler estimated parameters - sample mean, standard deviation

```python
# Using Perceptron Classification Algorithm for training of the model
from sklearn.linear_model import Perceptron
ppn = Perceptron(eta0=0.1, random_state=1)
ppn.fit(X_train_std, y_train)
```

eta0 - learning rate
n_iter - the num of epochs - passes over the training dataset

```python
y_pred = ppn.predict(X_test_std)
print('Misclassified examples: %d' % (y_test != y_pred).sum())
```

```python
# Classification accuracy of the perceptron on test dataset
from sklearn.metrics import accuracy_score
print('Accuracy: %.3f' % accuracy_score(y_test, y_pred))
```

```python
# Each classifier's prediction accuracy
print('Accuracy: %.3f' % ppn.score(X_test_std, y_test))
```

```python
from matplotlib.colors import ListedColormap
import matplotlib.pyplot as plt
def plot_decision_regions(X, y, classifier, test_idx=None, resolution=0.02):
    # setup marker generator & color map
    markers = ('s', 'x', 'o', '^', 'v')
    colors = ('red','blue','lightgreen','gray','cyan')
    cmap = ListedColormap(colors[:len(np.unique(y))])
    
    # plot the decision surface
    x1_min, x1_max = X[:, 0].min()-1, X[:, 0].max()+1
    x2_min, x2_max = X[:, 1].min()-1, X[:, 1].max()+1
    xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                          np.arange(x2_min, x2_max, resolution))
    Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
    Z = Z.reshape(xx1.shape)
    plt.contourf(xx1, xx2, Z, alpha=0.3, cmap=cmap)
    plt.xlim(xx1.min(), xx1.max())
    plt.ylim(xx2.min(), xx2.max())
    
    for idx, cl in enumerate(np.unique(y)):
        plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                   alpha=0.8, c=colors[idx],
                   marker=markers[idx], label=cl,
                   edgecolor='black')
        
    # highlight test examples
    if test_idx:
        # plot all examples
        X_test, y_test = X[test_idx, :], y[test_idx]
        
        plt.scatter(X_test[:, 0], X_test[:, 1],
                   c='', edgecolor='black', alpha=1.0,
                   linewidth=1, marker='o',
                   s=100, label='test set')
```

```python
X_combined_std = np.vstack((X_train_std, X_test_std))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X=X_combined_std,
                     y=y_combined,
                     classifier=ppn,
                     test_idx=range(105, 150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

As we can see from the above plot, the three flower classes cannot be separated by a linear decision boundary.

```python
# Using Logistic Regression for training 
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(C=100.0, random_state=1, solver='lbfgs', 
                        multi_class='ovr')
lr.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=lr, 
                      test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Predicting class labels
lr.predict(X_test_std[:3, :])
```

```python
# Using Support Vector Machines for training
from sklearn.svm import SVC
svm = SVC(kernel='linear', C=1.0, random_state=1)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std,
                     y_combined,
                     classifier=svm,
                     test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

As we can see linear logistic regression and linear SVM's often yield very similar results. Logistic regression tries to maximize the conditional likelihoods of the training data, which makes it more prone to outliers than SVM's, which mostly care about the points that are closest to the decision boundary. On the other hand, logistic regression has the advantage that it is a simpler model, can be implemented & updatedly easily, which is a bonus point while working with streaming data.

```python
# Using kernel SVM to solve nonlinear problem
# Radial Basis Function/Gaussian kernel
# C - inverse regularization param & here we can control the width of 
#     the margin & bias-variance tradeoff
svm = SVC(kernel='rbf', random_state=1, gamma=0.2, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std,
                     y_combined,
                     classifier=svm,
                     test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Notice the difference with large value of gamma
svm = SVC(kernel='rbf', random_state=1, gamma=100.0, C=1.0)
svm.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std,
                     y_combined,
                     classifier=svm,
                     test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

As we can see from the above resulting plot that the decision boundary around the classes 0 & 1 using a relatively large value of gamma.

```python
# Using Decision Tree - axis parallel decision boundaries
from sklearn.tree import DecisionTreeClassifier
tree_model = DecisionTreeClassifier(criterion='gini',max_depth=4,random_state=1)
tree_model.fit(X_train, y_train)
X_combined = np.vstack((X_train, X_test))
y_combined = np.hstack((y_train, y_test))
plot_decision_regions(X_combined, y_combined, classifier=tree_model, test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
from sklearn import tree
tree.plot_tree(tree_model)
plt.show()
```

Looking at the decision tree figure above, we can now nicely trace back the splits that the decision tree determined from our training dataset. We started with 105 examples at the root &  split them into 2 child nodes with 35 and 70 examples, using petal width cut-off <= 0.75cm. After the first split, we can see that the left child node is already pure & only contains examples from the Iris-setosa class. The further splits on the right are then used to separate the examples from the Iris-versicolor and Iris-virginica. Looking at this tree, and the decision region plot of the tree, we can see that the decision tree does a very good job of separating the flower classes.

```python
# Using Random Forest - ensemble of decision trees - for better generalization performance
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(criterion='gini',
                               n_estimators=25,
                               random_state=1,
                               n_jobs=2)
forest.fit(X_train, y_train)
plot_decision_regions(X_combined, y_combined,
                     classifier=forest, test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Using lazy learning(instance-based) - KNN 
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5, p=2, metric='minkowski')
knn.fit(X_train_std, y_train)
plot_decision_regions(X_combined_std, y_combined, classifier=knn, test_idx=range(105,150))
plt.xlabel('petal length [standardized]')
plt.ylabel('petal width [standardized]')
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
```

After looking into different ML algorithms, it is clear that the decision trees are particularly attractive if we take care about interpretability. On other hand logistic regression is not only a useful model for online learning via SGD, but also allows us to predict the probability of a particular event. Although SVMs are powerful linear models that can be extended to nonlinear problems via kernel trick, they've many parameters that have to be tuned to make good predictions. In contrast, ensemble methods, such as random forests, don't require much parameter tuning and don't overfit as easily as decision trees, which makes them more attractive models for many practical problem domains. The KNN classifier offers an alternative approach to classification via lazy learning that allows us to make predictions without any model training, but with a more computationally expensive prediction step.
